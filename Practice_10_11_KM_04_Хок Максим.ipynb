2. Програма має підраховувати біноміальні коефіцієнти для біномів усіх степенів, починаючи від 0 і закінчуючи тим, який ввів користувач.

3. Коефіцієнти мають виводитись у вигляді трикутника Паскаля.

Наприклад, якщо користувач введе число 5, програма має вивести результат:

1

1 1

1 2 1

1 3 3 1

1 4 6 4 1

1 5 10 10 5 1


```
def gen(x):
    r = 1
    a = [1,1]
    ls = [1,1]
    yield [1]
    while r <= x:
        yield ls
        ls = list()
        ls.append(1)
        n = 1
        for i in a[1:]:
            ls.append(i + a[n - 1])
            n += 1
        ls.append(1)
        a = ls
        r += 1



while True:
    try:
        x = int(input())

    except ValueError:
        print("Невірний тип даних")
        continue
    else:
        break


g = gen(x)
s = str()
for i in g:
    i = list(map(str,i))
    print(" ".join(i))
```
